import library yang dobutuhkan


```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

membaca gambar

```python
gambar=cv.imread("D:/semester 4/PCD/materi/PENGOLAHAN-CITRA-DIGITAL/aang.jpg")
plt.title("belajar translasi")
plt.imshow(gambar)
plt.show()
```

merubah ke grayscale

```python
gambar_gray=cv.cvtColor(gambar,cv.COLOR_BGR2GRAY)

plt.title("dirubah ke gray")
plt.imshow(gambar_gray,cmap="gray")
plt.show()
```

menampilkan histogramnya

```python
lebar,tinggi=gambar_gray.shape
hist=np.zeros(256)
print(gambar_gray)

for i in range(gambar_gray.shape[0]):
    for j in range (gambar_gray.shape[1]):
        hist[round(gambar_gray[i,j])]+=1
         
plt.plot(hist)
```
